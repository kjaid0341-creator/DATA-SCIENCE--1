```python
import numpy as np
a=input("enter").split()
ar=np.array(a,dtype=int)
print(ar)
val=np.sum(ar)
print("sum :",val)
val=np.mean(ar)
i = np.where(ar == val)
print("mean :",val, "on:",i[0]+1)
val=np.max(ar)
i = np.where(ar == val)
print("max :",np.max(ar),"on:",i[0]+1)
print("unique. score :",np.unique(ar))#give list of unique element
val=np.min(ar)
i = np.where(ar == val)
print("min :",val,"on:",i[0]+1)
print("std :",np.std(ar))
print("random score :",np.random.choice(ar))


```

```python
import numpy as np

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']  # Standard capitalization
scores = []

for i in range(7):
    # Use f-string with curly braces {}
    x = int(input(f"Enter the score for {days[i]} day: "))
    scores.append(x)

# Convert to NumPy array if needed
scores_array = np.array(scores)
print(scores_array)
```
