```python
# from os import rename
import pandas as pd
ddf=pd.read_csv("student_data.csv")
df1=pd.DataFrame({"id":[1,2,3,4],"name":["rewas","patil","moosa","andus"],"age":[25,30,35,40]})
df2=pd.DataFrame({"id":[3,4,5,6],"name":["moosa","andus","andusur","vadhusur"],"age":[25,30,35,40]})

df3=pd.merge(df1,df2, on="id", how='inner')#inner join
display(df3)
df3.rename(columns={'name_x':'name_df1','name_y':'name_df2'},inplace=True)#to rename the coloumn name
# inplace changes the orignal data frame bydefault False
display(df3)
df4=pd.merge(df1,df2, on="id", how='left')#left join
display(df4)
df5=pd.merge(df1,df2, on="id", how='right')#right join
display(df5)
df6=pd.merge(df1,df2, on="id", how='outer')#outer join
display(df6)

# cross join
df7=pd.merge(df1,df2, how='cross')#left join
display(df7)
```

```python
# concatination
df1=pd.DataFrame({"id":[3,4,5,6],"name":["moosa","andus","andusur","vadhusur"],"age":[25,30,35,40]})
df2=pd.DataFrame({"id":[1,2,3,4],"name":["rewas","patil","moosa","andus"],"age":[25,30,35,40]})
# pd.concat([df1,df2])
display(pd.concat([df1,df2]))
df3=pd.merge(df1,df2, on="id", how='inner')#inner join
display(df3)
```

```python
# drop dublicates
df=pd.DataFrame({"id":[3,4,5,6,6,3,3],"name":["moosa","andus","andusur","vadhusur","kalusur","niggasur","nanasur"],"age":[25,30,35,40,50,30,90]})
df1=df.drop_duplicates(subset=["id"],keep="first")
# display(df1
# keep-> if you want to keep last value or first
display(df1)
```
