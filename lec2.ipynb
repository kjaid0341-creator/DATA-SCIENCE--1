```python
lst=[1,2,3,4,5,'que',[2,4,8],'opaa']
print(lst)
#add single element in last
lst.append(20);
# extend() function add multiple elements in the list it take input as itrable data structure ex list to add elements at last
lst.extend([10,20,30,40])
print(lst)

#insert(index, value) add element on desired value
lst.insert(2,"booom")
print(lst)


#pop() it removes last element
lst.pop()
print(lst)

#remove(value) it removes value given to it , it removes only the first occurence
lst.remove('que')
print(lst)

#index() it returns index of the given value fist occurence
s=lst.index(20)
print(s)

#count() it returns number of value present in the list

l=lst.count(20)
print(l)

#sort() sort the list of only iniger value or only string value ,changes the list
h=[1,8,0,4,60]
h.sort()
print(h)

#sum ,min,max,sorted
print(min(h))
print(max(h))
print(sum(h))
print(sorted(h))
print(h)
```
